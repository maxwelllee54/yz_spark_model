```python
import pandas as pd

df = pd.read_excel('output_judy.xlsx')
df.head()
```

```python
goals = pd.read_csv('SDG_Goal.csv')
goals.columns = ['goal', 'text']

poverty = goals.iloc[0:5, :]
poverty
```

```python
from accessment_model import mySim
df_new = pd.DataFrame()
ind = 0

for i in poverty.index.values:
    target = poverty.loc[i, 'text']
    goal = str(poverty.loc[i, 'goal'])
    sentences = df.loc[df.loc[:, goal] > 0, 'p']
    for sentence in sentences:
        df_new.loc[sentence, goal ] = mySim(target, sentence)

```
